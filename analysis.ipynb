```python
import pandas as pd
import seaborn as sns
df_ti = pd.read_csv(r"D:\Adam Material\PROJECTS\Eleevo Pathways\TASK - Exploratory Data Analysis\Titanic.csv")
df_ti
```

```python
df_ti.isnull().sum() # check any column having null values
df_ti['Cabin'].fillna('Unknown', inplace=True)
df_ti['Embarked'].fillna('Unknown', inplace=True)
df_ti
```

```python
df_ti['Sex'] = df_ti['Sex'].astype('category')
df_ti['Pclass'] = df_ti['Pclass'].astype('category')
df_ti['Embarked'] = df_ti['Embarked'].astype('category')
df_ti['Survived'] = df_ti['Survived'].astype(int, errors='ignore')
df_ti['Age'] = df_ti['Age'].astype(int, errors='ignore')
df_ti.info() # show Dtype now for columns
```

```python
df_ti.groupby(['Sex', 'Pclass'])['Survived'].agg(['mean', 'count'])
```
